```python
import requests
from io import StringIO
import pandas as pd
```

```python
# URL of the dataset on GitHub
github_raw_url = 'https://raw.githubusercontent.com/mohandaz/HIA-303-Project/main/bca_train (1).csv'

# Fetch data from the URL
response = requests.get(github_raw_url)

if response.status_code == 200:
    # Load breast cancer data from CSV file using Pandas
    train= pd.read_csv(StringIO(response.text))

    # Print shape, columns, and the first few rows of the DataFrame
    print(train.shape)
    print(train.columns)
    train.head()
else:
    print(f"Failed to fetch data. Status Code: {response.status_code}")
```

```python
#drop PID column
train_2=train.drop(['PID'], axis= 1)
```

```python
# Dimenson reduction
#Apply low variance filter
var= train_2.var()
var
```

```python
train_var= pd.DataFrame({
    'columns':train_2.columns,
    'variance':var
})
train_var
```

```python
#Threshold= 0.1
columns_selected = train_var[train_var['variance']>0.1]['columns'].values
columns_selected
```

```python
#After applying low variance filter
train_selected = train_2[columns_selected]
train_selected.columns
#no column is below the threshold limit
```

```python
train_selected.head()
```

```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
```

```python
X= train_selected.drop(['Malignant'],axis=1)
y= train_selected['Malignant']
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 1)
```

```python
from scipy.stats import randint
from sklearn.model_selection import RandomizedSearchCV
param_dist = {'n_estimators':randint(50,500),
             'max_depth': randint(1,20)}
#Create a random forest classifier
rf= RandomForestClassifier()

#Use random search to find the best hyperparameters
rand_search = RandomizedSearchCV(rf,
                                 param_distributions= param_dist,
                                 n_iter= 5,
                                 cv=5)
#Fit the random search object to the data
rand_search.fit(X_train, y_train)
```

```python
#Create a variable for the best model
best_rf=  rand_search.best_estimator_

#Print the best hyperparameters
print('Best hyperparameters:', rand_search.best_params_)
```

```python
rf_all_features = RandomForestClassifier(random_state=1, n_estimators=226, max_depth=11)
rf_all_features.fit(X_train, y_train)
```

```python
accuracy_score(y_test, rf_all_features.predict(X_test))
```

```python
rfc= RandomForestClassifier(random_state=1, n_estimators= 226, max_depth=11)
boruta_selector = BorutaPy(rfc, n_estimators = 'auto', verbose= 2, random_state=1)
boruta_selector.fit(np.array(X_train), np.array(y_train))
```

```python
type(boruta_selector)
```

```python
print("Selected Features: ", boruta_selector.support_)    # check selected features
```

```python
print("Ranking: ",boruta_selector.ranking_)               # check ranking of features

print("No. of significant features: ", boruta_selector.n_features_)
```

```python
 #(The features with a ranking of 1 are selected).
selected_rf_features = pd.DataFrame({'Feature':list(X_train.columns),
                                      'Ranking':boruta_selector.ranking_})
selected_rf_features.sort_values(by='Ranking')
```

**Elastic Net**

```python
from sklearn.linear_model import ElasticNetCV, ElasticNet
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error
```

```python
#Determine best alpha parameter and l1 ratio
elastic_net= ElasticNet()
param_grid={"alpha":[0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.7, 1],
           "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 1]}
#perform gird search with cross-validation
grid_search = GridSearchCV(elastic_net, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

```python
# Train the model
e_net = ElasticNet(alpha = 0.001, l1_ratio=1)
e_net.fit(X_train, y_train)
```

```python
# calculate the prediction and mean square error
y_pred= e_net.predict(X_test)
mse= mean_squared_error(y_test, y_pred)
print("Mean Squared Error on test set", mse)
```

```python
e_net_coeff = pd.DataFrame()
e_net_coeff["Columns"] = X_train.columns
e_net_coeff['Coefficient Estimate'] = pd.Series(e_net.coef_)
e_net_coeff
```
